```python
# 바비 쇼핑몰 놀이세트('25')를 2개 이상 구매하면 청구 금액을 5% 할인
def discount(tran):
    if tran[3] == '25' and int(tran[4]) >= 2:
        tran[5] = str(round((float(tran[5]) * 0.95), 2))
    return tran
```

```python
transByCust = transByCust.mapValues(discount)
```

```python
import copy

# 사전('81')을 5개 이상 구매한 고객에게 사은품으로 칫솔('70') 보내기
def add_gift(tran):
    if tran[3] == '81' and int(tran[4]) >= 5:
        copied = copy.copy(tran)
        copied[3] = '70'
        copied[4] = '1'
        copied[5] = '0.00'
        return [tran, copied]
#         return [tran, [tran[0], tran[1], tran[2], '70', '1', '0.00']]
    else:
        return [tran]
```

```python
# transByCust = tranData.map(lambda tran: (int(tran[2]), tran))
```

```python
transByCust = transByCust.flatMapValues(add_gift)
```

```python
# 고객이 구매한 상품의 평균가격, 최저가격, 최고가격, 구매액 합계
def createComb(t):
    total = float(t[5])
    q = int(t[4])
    return (total/q, total/q, q, total)

def mergeVal(data, t):
    (mn, mx, c, tot) = data
    total = float(t[5])
    q = int(t[4])
    return (min(mn,total/q),max(mx,total/q),c+q,tot+total)

def mergeComb(data1, data2):
    (mn1, mx1, c1, tot1) = data1
    (mn2, mx2, c2, tot2) = data2
    return (min(mn1,mn1),max(mx1,mx2),c1+c2,tot1+tot2)
```

```python
transByCust.take(5)
```

```python
stats_by_cust = transByCust.combineByKey(createComb, mergeVal, mergeComb)
```
